```python
import pandas as pd 
from os.path import join
import numpy as np
# import MetaAnalysis_utils as mu
import seaborn as sns
import matplotlib.pyplot as plt
data_path = './processed'
output_path = './figures'

from matplotlib.pyplot import rc
import scipy.stats as stats 
sns.set_theme(style="white", font_scale=1.2,palette='viridis')
font = {'weight' : 'bold', 'family':'Nimbus Sans'}
rc('font', **font) 
```

***Hit rate box plot with dots***

```python
hannah = pd.read_csv(join(data_path, f'Hannah.csv'), index_col = 0)
hannah['Study'] = 'Hannah'
martin = pd.read_csv(join(data_path, 'Martin.csv'), index_col = 0)
martin['Study'] = 'Martin'
```

```python
# Assume data for both plots are stored in data1 and data2
order = ['Synonymous', 'Missense', 'No-Edits','Non-coding']

fig, axes = plt.subplots(1, 2, figsize=(10, 5), sharey=True)
sns.boxplot(data=hannah, x='mutation_category', order=order, y='perc',
            boxprops={'alpha': 0.4}, ax=axes[0],linewidth=2)
sns.stripplot(data=hannah, x='mutation_category', y='perc', dodge=True, size=8,hue="cell_line",
              order=order, ax=axes[0])
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=30)
axes[0].set_ylabel('Hit Rate (%)',fontweight='bold')
axes[0].set_xlabel('')
axes[0].set_title('Hannah et al',fontweight='bold')
axes[0].spines['top'].set_visible(False)
axes[0].spines['right'].set_visible(False)
axes[0].legend(title="Cell Line",  bbox_to_anchor = (1,1),loc='center left', fontsize='xx-small')

sns.boxplot(data=martin, x='mutation_category', order=order, y='perc',
            boxprops={'alpha': 0.4}, ax=axes[1],linewidth=2.5)
sns.stripplot(data=martin, x='mutation_category', y='perc', dodge=True, size=8,
              order=order, ax=axes[1], hue = 'cell_line')
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=30)
axes[1].set_xlabel('')
axes[1].set_title('Cuella-Martin et al',fontweight='bold')
axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)
axes[1].spines['left'].set_visible(False)
# axes[1].legend(title="Cell Line", bbox_to_anchor=(1, 1), loc='lower right', fontsize='x-small')
axes[1].set_ylabel('')
axes[1].legend(title="Cell Line",  bbox_to_anchor = (1,1),loc='center left', fontsize='xx-small')


fig.suptitle('Reported Hit Rate', fontweight='bold')

# Adjust layout
plt.tight_layout()
plt.savefig(join('./figures/report_hit.png'))
plt.show()
```
