```python
import numpy as np
import pandas as pd
import seaborn as sns
from os.path import join

import matplotlib.pyplot as plt
from matplotlib.pyplot import rc

output_path = './Figures'
data_path  = './processed'

sns.set_theme(style="white", font_scale=2,palette='viridis')
font = {'weight' : 'bold', 'family':'Nimbus Sans'}
rc('font', **font) 
```

```python
# Data from the provided plot
original_positions = np.arange(1, 16)
original_values = [0.002, 0.01, 0.05, 0.14, 0.17, 0.17, 0.16, 0.13, 0.06, 0.04, 0.03, 0.02, 0.001, 0.00, 0.00]

# Extend the x-axis to the range -3 to 23
extended_positions = np.arange(1, 24)
extended_values = np.zeros_like(extended_positions, dtype=float)

# Map the original data to the corresponding positions in the extended range
for pos, val in zip(original_positions, original_values):
    extended_values[pos-1] = val  # Adjusting for the extended range (-3 to 23)

# Plot the mock data
plt.figure(figsize=(15, 4))
plt.bar(extended_positions, extended_values, color='brown', alpha=0.8, edgecolor='black')

# Customize the plot

# Customize the plot
plt.xticks(extended_positions, fontsize=10, rotation=90)  # Adjust font size and rotation for alignment
plt.yticks(fontsize=10)
plt.legend(fontsize=10)
plt.title("Editing Efficiency", fontweight = 'bold')


# Remove the top and right spines
for spine in ['top', 'right', 'left']:
    plt.gca().spines[spine].set_visible(False)
plt.gca().spines['bottom'].set_position(('outward', 10))  # Move bottom spine outward for better visibility
plt.tight_layout()

plt.savefig('./Figures/fig1_editing_efficiency.png')
```
